```r
suppressWarnings(suppressPackageStartupMessages({
    library(tidyverse)
    library(data.table)
}))
```

```r
df <- fread('8_annotate_eval_df.tsv')
```

```r
df %>%
filter(geno_delta > 0.01, split == 'white_british:test') -> df_sub
```

```r
options(repr.plot.width=4, repr.plot.height=4)
df_sub %>%
ggplot(aes(x=N_GBE,y=n_variables))+
geom_point() + theme_bw()
```

```r
cor.test(df_sub$n_variables, df_sub$N_GBE, method='spearman')
```

```r
options(repr.plot.width=4, repr.plot.height=4)
df_sub %>%
ggplot(aes(x=n_variables,y=geno_delta))+
geom_point() + theme_bw()
```

```r
cor.test(df_sub$n_variables, df_sub$geno_delta, method='spearman')
```

```r
cor.test(df_sub$N_GBE, df_sub$geno_delta, method='spearman')
```
