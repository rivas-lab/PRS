```r
require(tidyverse)
require(data.table)
```

```r
snpnet_scores_f <- '~/repos/rivas-lab/PRS/public_output/snpnet_PRS/cancer_20190303.tsv'
```

```r
snpnet_scores_df <- fread(
    cmd=paste('cat', snpnet_scores_f, '| sed -e "s/^#//g"', sep=' '), sep='\t'
)
```

```r
snpnet_scores_df %>% 
rename(AUC = R_or_AUC) %>%
filter(
    N > 1000
#     phe_name %in% c(
#         'lung_cancer', 'breast_cancer', 'bladder_cancer'
#     )
) %>%
spread(features, AUC) %>%
mutate(
    delta = Genotype_and_covariates - Covariates_only,
    rank_geno = rank(-Genotype_only)
) %>%
arrange(-delta)
```

```r
vanilla_scores_f <- '~/repos/rivas-lab/PRS/public_output/vanilla_PRS/cancer_20190302.tsv'
```

```r
vanilla_scores_df <- fread(
    cmd=paste('cat', vanilla_scores_f, '| sed -e "s/^#//g"', sep=' '), sep='\t'
)
```

```r
vanilla_scores_df %>% 
rename(AUC = R_or_AUC) %>%
filter(
    N > 1000
#     phe_name %in% c(
#         'lung_cancer', 'breast_cancer', 'bladder_cancer'
#     )
) %>%
spread(features, AUC) %>%
mutate(delta = PRS_and_covars - covars) %>%
group_by(GBE_ID) %>%
mutate(
    rank = rank(-delta)
) %>%
ungroup()%>%
filter(rank == 1) %>%
mutate(
    rank_geno = rank(-PRS)
) %>%
arrange(-delta)
```
