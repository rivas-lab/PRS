```r
require(tidyverse)
require(data.table)
```

```r
var_anno <- '/oak/stanford/groups/mrivas/private_data/ukbb/variant_filtering/variant_filter_table.tsv.gz'
var_df <- fread(cmd=paste0('zcat ', var_anno))
```

```r
dff <- var_df %>% filter(Gene_symbol == 'CTLA4')
```

```r
dff
```

```r
gwas_WB_f  <- '/oak/stanford/groups/mrivas/ukbb24983/cal/gwas/extras/cancer/white_british/ukb24983_v2_hg19.cancer1060.genotyped.PHENO1.glm.logistic.hybrid.gz'
gwas_NBW_f <- '/oak/stanford/groups/mrivas/ukbb24983/cal/gwas/extras/cancer/non_british_white/ukb24983_v2_hg19.cancer1060.genotyped.glm.logistic.hybrid.gz'

gwas_WB_df  <- fread(cmd=paste0('zcat ', gwas_WB_f)) %>% rename('CHROM' = '#CHROM')
gwas_NBW_df <- fread(cmd=paste0('zcat ', gwas_NBW_f)) %>% rename('CHROM' = '#CHROM')
```

```r
gwas_WB_df %>% 
inner_join(
    dff,
    by=c('CHROM', 'POS', 'REF', 'ALT', 'ID')
) %>%
fwrite(
    'white_british.ukb24983_v2_hg19.cancer1060.genotyped.glm.logistic.hybrid.tsv',
    sep='\t'
)
```

```r
gwas_NBW_df %>% 
inner_join(
    dff,
    by=c('CHROM', 'POS', 'REF', 'ALT', 'ID')
) %>%
fwrite(
    'non_british_white.ukb24983_v2_hg19.cancer1060.genotyped.glm.logistic.hybrid.tsv',
    sep='\t'
)
```
